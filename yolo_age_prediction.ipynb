```python
import os
import shutil
import random
from tqdm import tqdm

# 📁 Mevcut klasörler
formatted_image_dir = "/content/data/FairFace/train"  # Yeniden adlandırılmış img klasörü
output_base_dir = "/content/fairface_yolo"

# 🔧 Sadece test ayırma oranı
TEST_SPLIT = 0.05

# 🗂️ Klasör yapısını oluştur (val yok)
os.makedirs(os.path.join(output_base_dir, 'train', 'images'), exist_ok=True)
os.makedirs(os.path.join(output_base_dir, 'train', 'labels'), exist_ok=True)
os.makedirs(os.path.join(output_base_dir, 'test', 'images'), exist_ok=True)
os.makedirs(os.path.join(output_base_dir, 'test', 'labels'), exist_ok=True)

# 🔀 Tüm resimleri al
all_images = [f for f in os.listdir(formatted_image_dir) if f.endswith('.jpg')]
random.shuffle(all_images)

test_size = int(TEST_SPLIT * len(all_images))
test_images = all_images[:test_size]
train_images = all_images[test_size:]

# 💾 Label üretme fonksiyonu
def generate_label_line(filename):
    parts = filename.split('_')
    if len(parts) < 4:
        return None

    gender = int(parts[0])
    age = int(parts[1])
    race = int(parts[2])

    yolo_box = "0.5 0.5 1.0 1.0"
    return f"0 {yolo_box} age:{age} gender:{gender} race:{race}"

# 🚀 Taşı ve label oluştur
def process_split(split, image_list):
    for img_file in tqdm(image_list, desc=f"{split.upper()} verisi işleniyor"):
        # Kopyala
        src = os.path.join(formatted_image_dir, img_file)
        dst_img = os.path.join(output_base_dir, split, 'images', img_file)
        shutil.copy(src, dst_img)

        # Etiket dosyası
        label_line = generate_label_line(img_file)
        if label_line:
            label_name = os.path.splitext(img_file)[0] + ".txt"
            label_path = os.path.join(output_base_dir, split, 'labels', label_name)
            with open(label_path, 'w') as f:
                f.write(label_line + "\n")

# 🔁 İşlemleri başlat
process_split("train", train_images)
process_split("test", test_images)
```

```python
import os
import shutil
from tqdm import tqdm

# 🔧 Giriş ve çıkış dizinleri
source_base = "/content/fairface_yolo"
output_base = "/content/age_yolo"  # sadece age etiketli dataset

splits = ["train", "val", "test"]

# 🗂️ Çıktı klasörlerini oluştur
for split in splits:
    os.makedirs(os.path.join(output_base, split, "images"), exist_ok=True)
    os.makedirs(os.path.join(output_base, split, "labels"), exist_ok=True)

# ✅ Age etiketini çıkaran fonksiyon
def extract_age_label(label_path):
    with open(label_path, 'r') as f:
        line = f.readline().strip()
        if "age:" in line:
            try:
                age_part = line.split("age:")[1].split()[0]
                age_class = int(age_part)
                return f"{age_class} 0.5 0.5 1.0 1.0"
            except:
                return None
    return None

# 🔁 Splitleri işle
for split in splits:
    print(f"🔄 {split.upper()} seti işleniyor...")
    img_dir = os.path.join(source_base, split, "images")
    label_dir = os.path.join(source_base, split, "labels")

    output_img_dir = os.path.join(output_base, split, "images")
    output_label_dir = os.path.join(output_base, split, "labels")

    image_files = [f for f in os.listdir(img_dir) if f.endswith(".jpg")]

    for img_file in tqdm(image_files):
        label_file = img_file.replace(".jpg", ".txt")
        label_path = os.path.join(label_dir, label_file)

        if not os.path.exists(label_path):
            continue

        age_label_line = extract_age_label(label_path)

        if age_label_line is not None:
            # Görseli kopyala
            shutil.copy(os.path.join(img_dir, img_file),
                        os.path.join(output_img_dir, img_file))

            # Yeni label dosyasını yaz
            with open(os.path.join(output_label_dir, label_file), 'w') as f:
                f.write(age_label_line + "\n")
```

```python
# 📌 Google Drive'ı bağla
!pip install -q ultralytics
from google.colab import drive
import os
import shutil
from ultralytics import YOLO
from IPython.display import display, Javascript


# 📁 Eğitim klasörü
drive_project_dir = "/content/drive/MyDrive/2209yoloAgeV2"
model_name = "age_model"
model_dir = os.path.join(drive_project_dir, model_name)

# 🚀 YOLO modeliyle eğitim (augmentations kapalı, m modeli)
!yolo train \
  task=detect \
  model=yolo11m.pt \
  data=/content/age_yolo/data.yaml \
  epochs=50 \
  batch=270 \
  imgsz=640 \
  cfg=/content/age_yolo/no_aug.yaml \
  plots=True \
  project={drive_project_dir} \
  name={model_name}

# ✅ Model eğitildikten sonra ağırlık dosyasını yükle
model_path = os.path.join(model_dir, "weights", "best.pt")

# 📌 Dosya gerçekten oluşmuş mu kontrol et
if os.path.exists(model_path):
    model = YOLO(model_path)

    # 🧪 Test setinde doğrulama ve confusion matrix oluşturma
    metrics = model.val(
        data="/content/age_yolo/data.yaml",
        split="test",
        plots=True,
        save_json=True
    )

    # 📁 Confusion matrix görselini Google Drive'a taşı
    yolo_val_dir = metrics.save_dir
    drive_eval_dir = os.path.join(model_dir, "eval")
    os.makedirs(drive_eval_dir, exist_ok=True)

    for file_name in os.listdir(yolo_val_dir):
        if "confusion_matrix" in file_name and file_name.endswith(".png"):
            full_path = os.path.join(yolo_val_dir, file_name)
            shutil.copy(full_path, os.path.join(drive_eval_dir, file_name))

    print("✅ Eğitim tamamlandı, ağırlıklar ve confusion matrix Drive'a kaydedildi.")

    # 📴 Çalışma zamanını kapat (GPU bırakılır)
    print("📴 Çalışma zamanı sonlandırılıyor...")
    display(Javascript('''google.colab.kernel.invokeFunction("notebook.shutdown", [], {})'''))

else:
    print("❌ Eğitim başarısız oldu, 'best.pt' dosyası bulunamadı.")
```
